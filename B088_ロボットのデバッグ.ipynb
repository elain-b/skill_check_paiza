```
H,W=map(int,input().split())
muki=1
x,y=0,0
turn=0
s=[]
def func():
    global turn,muki
    if turn>=2:
        return False
    else:
        muki=(muki+turn+1)%4
        turn+=1
    return True
for h in range(H):
    s.append(list(input()))
while True:
    if muki==1:
        if x==W-1:
            if not func():
                break
        elif s[y][x+1]=='.':
            s[y][x]='#'
            x+=1
            turn=0
        elif s[y][x+1]=='#':
            if not func():
                break
    elif muki==2:
        if y==H-1:
            if not func():
                break
        elif s[y+1][x]=='.':
            s[y][x]='#'
            y+=1
            turn=0
        elif s[y+1][x]=='#':
            if not func():
                break
    elif muki==3:
        if x==0:
            if not func():
                break
        elif s[y][x-1]=='.':
            s[y][x]='#'
            x-=1
            turn=0
        elif s[y][x-1]=='#':
            if not func():
                break
    elif muki==0:
        if y==0:
            if not func():
                break
        if s[y-1][x]=='.':
            s[y][x]='#'
            y-=1
            turn=0
        elif s[y-1][x]=='#':
            if not func():
                break
print(x,y)
```

```
H,W=map(int,input().split())
muki=1
x,y=1,1
turn=0
s=[]
def func():
    global turn,muki
    if turn>=2:
        return False
    else:
        muki=(muki+turn+1)%4
        turn+=1
    return True
s.insert(0,['#' for i in range(W+2)])
for h in range(H):
    s.append(list(input()))
    s[h+1].insert(0,'#')
    s[h+1].append('#')
s.append(['#' for i in range(W+2)])
while True:
# for i in range(3):
    if muki==1:
        if s[y][x+1]=='.':
            s[y][x]='#'
            x+=1
            turn=0
        elif s[y][x+1]=='#':
            if not func():
                break
    elif muki==2:
        if s[y+1][x]=='.':
            s[y][x]='#'
            y+=1
            turn=0
        elif s[y+1][x]=='#':
            if not func():
                break
    elif muki==3:
        if s[y][x-1]=='.':
            s[y][x]='#'
            x-=1
            turn=0
        elif s[y][x-1]=='#':
            if not func():
                break
    elif muki==0:
        if s[y-1][x]=='.':
            s[y][x]='#'
            y-=1
            turn=0
        elif s[y-1][x]=='#':
            if not func():
                break
# print(s)
print(muki,turn)
print(x-1,y-1)
```
